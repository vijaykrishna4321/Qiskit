```python
from qiskit import*
from qiskit.visualization import plot_bloch_multivector, visualize_transition
from qiskit.visualization import plot_histogram
%matplotlib inline 
import math
```

```python
#Create a quantum circuit with a three qubit
#the default initial state of qubit will be |0> or [1,0]
qc = QuantumCircuit(3)
```

```python
plot_bloch_multivector (qc)
```

```python
#Get the backend for the circuit (simulator or realtime system)
backend = Aer.get_backend('statevector_simulator')
```

```python
#Apply the pauli X gate
qc.h(0)
qc.h(1)
qc.h(2)
```

```python
#Draw the circuit
#qc.draw()
qc.draw('mpl')
```

```python
#Get the backend for the circuit(simulator or real time system)
backend = Aer.get_backend('statevector_simulator')
```

```python
plot_bloch_multivector (qc)
```

```python
#execute the circuit using the backend
out = execute(qc,backend).result().get_statevector()
```

```python
from qiskit_textbook.tools import array_to_latex
array_to_latex(out,pretext= "\\text{statevector}= ")
```

```python
#Plot the result as bloch sphere visulization
plot_bloch_multivector(out)
```

```python
result = execute(qc,backend).result()
counts = result.get_counts()
plot_histogram(counts)
```
